### توضیح کلی داده‌ها (دیتافریم)

شما یک مجموعه داده (dataset) بسیار غنی و دقیق از **۱۰۰۰۰ واحد مسکونی (آپارتمان)** دارید. این دیتافریم شامل **۴۸ ویژگی (ستون)** است که جزئیات بسیار کاملی از مشخصات فنی، تجهیزات و مصالح به کار رفته در ساختمان‌ها را ارائه می‌دهد. هدف نهایی، پیش‌بینی **قیمت (`Price`)** هر واحد بر اساس این ویژگی‌ها است.

از بین این ۴۸ ستون:
*   **۶ ستون** از نوع عددی (`int64`) هستند.
*   **۴۲ ستون** از نوع رسته‌ای (`object` یا رشته‌ای) هستند.

همانطور که از خروجی `nunique()` مشخص است، بیشتر متغیرهای رسته‌ای، تعداد کمی کلاس منحصر به فرد (عمدتاً ۲ یا ۳ گزینه) دارند که نشان‌دهنده‌ی استانداردها و پکیج‌های مختلف کیفی (مثلاً اقتصادی، استاندارد، لوکس) در ساخت‌وساز است. متغیر `Price` (قیمت) تنوع بسیار بالایی (`8794` مقدار منحصر به فرد از ۱۰۰۰۰ نمونه) دارد که نشان‌دهنده یک مسئله **رگرسیون** است.

```python
# وارد کردن کلاس StandardScaler برای استانداردسازی ویژگی‌های عددی
from sklearn.preprocessing import StandardScaler

# وارد کردن کلاس OneHotEncoder برای کدگذاری ویژگی‌های دسته‌ای
from sklearn.preprocessing import OneHotEncoder

# وارد کردن کلاس ColumnTransformer برای ایجاد یک ترانسفورمر pipeline بر اساس نوع ستون‌ها
from sklearn.compose import ColumnTransformer
```

```python
# تعریف لیست ستون‌های عددی برای پردازش
numerical_cols = ['Floor', 'Unit_Area_sq_m', 'Number_of_Rooms', 'Parking_Spots', 'Storage_Unit']

# ایجاد لیست ستون‌های دسته‌ایی با انتخاب تمام ستون‌هایی که در لیست numerical_cols وجود ندارند
categorical_cols = [col for col in X.columns if col not in numerical_cols]
```

```python
# ایجاد یک transformer برای ویژگی‌های عددی با استفاده از StandardScaler برای استانداردسازی داده‌های عددی
numerical_transformer = StandardScaler()

# ایجاد یک transformer برای ویژگی‌های دسته‌ای با استفاده از OneHotEncoder
# handle_unknown='ignore': در صورت مواجهه با دسته‌های جدید در داده آزمون، آن‌ها را نادیده می‌گیرد
# sparse_output=False: خروجی را به صورت ماتریس متراکم (dense) برمی‌گرداند نه ماتریس پراکنده (sparse)
categorical_transformer = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
```

```python
# ایجاد یک پیش‌پردازشگر ترکیبی برای پردازش ستون‌های مختلف
preprocessor = ColumnTransformer(
    transformers=[
        # پردازشگر برای ویژگی‌های عددی: استفاده از numerical_transformer برای ستون‌های numerical_cols
        ('num', numerical_transformer, numerical_cols),
        # پردازشگر برای ویژگی‌های دسته‌ای: استفاده از categorical_transformer برای ستون‌های categorical_cols
        ('cat', categorical_transformer, categorical_cols)
    ])
```

```python
# انجام پیش‌پردازش و استانداردسازی روی داده‌های آموزشی

X_train_encoded = pd.DataFrame(
    preprocessor.fit_transform(X_train),  # آموزش پیش‌پردازشگر و تبدیل داده‌های آموزشی
    columns=preprocessor.get_feature_names_out(),  # گرفتن نام ستون‌های پس از تبدیل
    index=X_train.index  # حفظ ایندکس اصلی داده‌های آموزشی
)

# تبدیل داده‌های اعتبارسنجی با استفاده از پیش‌پردازشگر آموزش دیده
X_val_encoded = pd.DataFrame(
    preprocessor.transform(X_val),  # تبدیل داده‌های اعتبارسنجی (بدون آموزش مجدد)
    columns=preprocessor.get_feature_names_out(),  # گرفتن نام ستون‌های پس از تبدیل
    index=X_val.index  # حفظ ایندکس اصلی داده‌های اعتبارسنجی
)

# تبدیل داده‌های آزمون با استفاده از پیش‌پردازشگر آموزش دیده
X_test_encoded = pd.DataFrame(
    preprocessor.transform(X_test),  # تبدیل داده‌های آزمون (بدون آموزش مجدد)
    columns=preprocessor.get_feature_names_out(),  # گرفتن نام ستون‌های پس از تبدیل
    index=X_test.index  # حفظ ایندکس اصلی داده‌های آزمون
)
```

```python
# نمایش 5 سطر اول از دیتافریم X_train_encoded
# این دیتافریم شامل داده‌های آموزشی پس از پیش‌پردازش و استانداردسازی می‌باشد
X_train_encoded.head()
```

```python
# استانداردسازی متغیر هدف (y) برای داده‌های آموزشی
# تبدیل y_train به DataFrame و سپس اعمال استانداردسازی با استفاده از numerical_transformer
y_train_encoded = numerical_transformer.fit_transform(y_train.to_frame())

# استانداردسازی متغیر هدف (y) برای داده‌های اعتبارسنجی
# تبدیل y_val به DataFrame و سپس اعمال تبدیل با استفاده از numerical_transformer آموزش دیده
y_val_encoded = numerical_transformer.transform(y_val.to_frame())

# استانداردسازی متغیر هدف (y) برای داده‌های آزمون
# تبدیل y_test به DataFrame و سپس اعمال تبدیل با استفاده از numerical_transformer آموزش دیده
y_test_encoded = numerical_transformer.transform(y_test.to_frame())
```

# ساخت شبکه مصنوعی

```python
# وارد کردن کتابخانه‌های لازم برای ساخت شبکه عصبی
import tensorflow.keras as keras
from tensorflow.keras import regularizers, initializers
from tensorflow.keras import backend as K

# تعریف تابع محاسبه ضریب تعیین (R-squared) به عنوان معیار ارزیابی سفارشی
def r_squared(y_true, y_pred):
    """
    محاسبه ضریب تعیین (R-squared) به عنوان معیار ارزیابی
    """
    SS_res = K.sum(K.square(y_true - y_pred))  # محاسبه مجموع مربعات باقیمانده
    SS_tot = K.sum(K.square(y_true - K.mean(y_true)))  # محاسبه مجموع مربعات کل
    return 1 - SS_res/(SS_tot + K.epsilon())  # محاسبه R-squared با افزودن epsilon برای جلوگیری از تقسیم بر صفر

# تعریف تابع ساخت مدل شبکه عصبی
def build_model():
    """
    مدل ANN ساده با مقدار متوسطی از dropout و regularization (L2).
    """
    # تعریف لایه پنهان اول با 64 نرون، تابع فعال‌ساز relu و تنظیمات initializer و regularizer
    hidden_layer = keras.layers.Dense(units=64, activation="relu",input_shape=(127,),kernel_initializer=initializers.HeNormal(),kernel_regularizer=regularizers.l2(0.001))
    
    # تعریف لایه پنهان دوم با 32 نرون و تنظیمات مشابه
    hidden_layer2 = keras.layers.Dense(units=32, activation="relu",kernel_initializer=initializers.HeNormal(),kernel_regularizer=regularizers.l2(0.001))

    # تعریف لایه خروجی با 1 نرون و تابع فعال‌ساز خطی برای مسئله رگرسیون
    output_layer = keras.layers.Dense(units=1, activation="linear",kernel_initializer='glorot_uniform' )

    # ایجاد مدل Sequential
    model = keras.models.Sequential()

    # افزودن لایه پنهان اول به مدل
    model.add(hidden_layer)
    # افزودن لایه Dropout با نرخ 0.1 برای کاهش overfitting
    model.add(keras.layers.Dropout(0.1))
    # افزودن لایه پنهان دوم به مدل
    model.add(hidden_layer2)
    # افزودن لایه Dropout دوم با نرخ 0.1
    model.add(keras.layers.Dropout(0.1))

    # افزودن لایه خروجی به مدل
    model.add(output_layer)

    # کامپایل مدل با تنظیمات بهینه‌ساز، تابع خطا و معیارهای ارزیابی
    model.compile(
        optimizer='adam',  # استفاده از بهینه‌ساز Adam
        loss='mse',  # تابع خطای میانگین مربعات خطا
        metrics=['mae', 'mse',r_squared]  # معیارهای ارزیابی: خطای مطلق میانگین، میانگین مربعات خطا و R-squared
    )
    return model

# ایجاد مدل با فراخوانی تابع build_model
model=build_model()
```

```python
# نمایش خلاصه‌ای از معماری مدل شبکه عصبی
# این متد اطلاعاتی درباره لایه‌ها، تعداد پارامترها و شکل خروجی هر لایه نمایش می‌دهد
model.summary()
```

```python
# وارد کردن کلاس‌های callback لازم از کتابخانه Keras
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# تعریف EarlyStopping برای توقف زودهنگام آموزش وقتی که validation_loss بهبود نمی‌یابد
early = EarlyStopping(
    monitor='val_loss',  # نظارت بر روی loss داده‌های اعتبارسنجی
    patience=5,  # تعداد epochهای بدون بهبود قبل از توقف آموزش
    restore_best_weights=True  # بازیابی بهترین وزن‌های مدل پس از توقف
)

# آموزش مدل شبکه عصبی با پارامترهای مشخص
history = model.fit(
    X_train_encoded,  # داده‌های آموزشی پس از پیش‌پردازش
    y_train_encoded,  # برچسب‌های آموزشی پس از پیش‌پردازش
    epochs=100,  # حداکثر تعداد epochهای آموزش
    batch_size=32,  # اندازه batch برای آموزش
    validation_data=(X_val_encoded, y_val_encoded),  # داده‌های اعتبارسنجی برای مانیتورینگ

    # استفاده از callback های تعریف شده
    callbacks=[
        early,  # callback برای توقف زودهنگام
        ReduceLROnPlateau(  # callback برای کاهش نرخ یادگیری هنگام توقف بهبود
            monitor='val_loss',  # نظارت بر روی loss داده‌های اعتبارسنجی
            factor=0.2,  # ضریب کاهش نرخ یادگیری (ضرب در 0.2)
            patience=5,  # تعداد epochهای بدون بهبود قبل از کاهش نرخ یادگیری
            min_lr=1e-6  # حداقل مقدار نرخ یادگیری
        )
    ],
    verbose=1  # نمایش پیشرفت آموزش (1 = نمایش نوار پیشرفت)
)
```

```python
# انجام پیش‌بینی روی داده‌های آزمون با استفاده از مدل آموزش دیده
y_pred = model.predict(X_test_encoded)

# ارزیابی مدل روی داده‌های آموزشی و ذخیره نتایج
train_results = model.evaluate(X_train_encoded, y_train_encoded)
print("Evaluation results train:")
# چاپ معیارهای ارزیابی برای داده‌های آموزشی
for name, value in zip(model.metrics_names, train_results):
    print(f"train_{name}: {value:.4f}")

print(100*"*")  # چاپ خط جداکننده

# ارزیابی مدل روی داده‌های اعتبارسنجی و ذخیره نتایج
val_results = model.evaluate(X_val_encoded, y_val_encoded)
print("Evaluation results val:")
# چاپ معیارهای ارزیابی برای داده‌های اعتبارسنجی
for name, value in zip(model.metrics_names, val_results):
    print(f"test_{name}: {value:.4f}")

print(100*"*")  # چاپ خط جداکننده

# ارزیابی مدل روی داده‌های آزمون و ذخیره نتایج
test_results = model.evaluate(X_test_encoded, y_test_encoded)
print("Evaluation results test:")
# چاپ معیارهای ارزیابی برای داده‌های آزمون
for name, value in zip(model.metrics_names, test_results):
    print(f"test_{name}: {value:.4f}")

# بازگرداندن پیش‌بینی‌ها به مقیاس اصلی با استفاده از تابع inverse_transform
y_pred_original = numerical_transformer.inverse_transform(y_pred)
# بازگرداندن مقادیر واقعی به مقیاس اصلی (با reshape برای تطابق شکل)
y_test_original = numerical_transformer.inverse_transform(y_test_encoded.reshape(-1, 1))

# نمایش نمونه‌هایی از پیش‌بینی در مقیاس اصلی
print("\nنمونه‌هایی از پیش‌بینی در مقیاس اصلی:")
# چاپ 10 نمونه اول پیش‌بینی‌ها (به صورت گرد شده به 2 رقم اعشار)
print("Predicted (Original):", y_pred_original[:10].flatten().round(2))
# چاپ 10 نمونه اول مقادیر واقعی (به صورت گرد شده به 2 رقم اعشار)
print("Actual (Original):", y_test_original[:10].flatten().round(2))
```
